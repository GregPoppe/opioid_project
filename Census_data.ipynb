```python
#Create function to request json from Census API and return a DataFrame

def data_to_df(base_url,params):
    response = requests.get(base_url, params=params)
    data = response.json()
    df = pd.DataFrame.from_records(data[1:])
    df.columns = data[0] 
    return df
```
